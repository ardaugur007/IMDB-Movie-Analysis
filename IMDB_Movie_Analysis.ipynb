```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Guncel ve calisan yeni veri seti linki
url = "https://raw.githubusercontent.com/peetck/IMDB-Top1000-Movies/master/IMDB-Movie-Data.csv"
df = pd.read_csv(url)

display(df.head())
print(df.info())

df.dropna(inplace=True)

plt.figure(figsize=(10, 5))
yearly_runtime = df.groupby('Year')['Runtime (Minutes)'].mean()
sns.lineplot(x=yearly_runtime.index, y=yearly_runtime.values, marker='o', color='b')
plt.title('Average Runtime Per Year')
plt.ylabel('Minutes')
plt.show()

plt.figure(figsize=(12, 6))
top_directors = df.groupby('Director')['Votes'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_directors.values, y=top_directors.index, hue=top_directors.index, palette='viridis', legend=False)
plt.title('Top 10 Directors by Total Votes')
plt.xlabel('Total Votes')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rating', y='Revenue (Millions)', data=df, alpha=0.6, color='r')
plt.title('Rating vs Revenue')
plt.show()
```
